```python
def update_gender(row):
    if row['Gender'] == 'Not specified':
        if row['Title'] in ['Mr']:
            return 'Male'
        elif row['Title'] in ['Ms', 'Mrs', 'Miss']:
            return 'Female'
    return row['Gender'] 

data['Gender'] = data.apply(update_gender, axis=1)

```

```python
data[numeric_columns].hist(bins=20, figsize=(10, 8))
plt.suptitle('Distribution of Numerical Variables', fontsize=16)
plt.show()
```

```python
categorical_columns = ['LegalType','IsVATRegistered','Bank','AccountType','MaritalStatus','Gender']

for column in categorical_columns:
    data[column].value_counts().plot(kind='bar', figsize=(6, 4),title=f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
categorical_columns = ['Province','MainCrestaZone','SubCrestaZone']
for column in categorical_columns:
    data[column].value_counts().plot(kind='bar', figsize=(6, 4), title=f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
categorical_columns = ['VehicleType','make','AlarmImmobiliser','TrackingDevice','NewVehicle','WrittenOff','Rebuilt']
for column in categorical_columns:
    data[column].value_counts().plot(kind='bar', figsize=(6, 4), title=f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
categorical_columns = ['TermFrequency','CoverCategory','CoverType','CoverGroup','Section','Product']
for column in categorical_columns:
    data[column].value_counts().plot(kind='bar', figsize=(6, 4), title=f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
grouped_data = data.groupby('PostalCode').agg({
    'TotalPremium': 'mean',
    'TotalClaims': 'mean'
}).reset_index()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(grouped_data['TotalPremium'], grouped_data['TotalClaims'], alpha=0.5)
plt.title('TotalPremium vs TotalClaims by ZipCode')
plt.xlabel('Average TotalPremium')
plt.ylabel('Average TotalClaims')
plt.grid(True)
plt.show()
```

```python
correlation_matrix = grouped_data[['TotalPremium', 'TotalClaims']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
cover_by_geo = data.groupby(['Province', 'CoverType']).size().unstack()

cover_by_geo.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Comparison of Insurance Cover Types by State')
plt.ylabel('Count')
plt.xlabel('State')
plt.legend(title='Cover Type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
premium_by_geo = data.groupby('Province')['TotalPremium'].mean().sort_values()

plt.figure(figsize=(10, 6))
premium_by_geo.plot(kind='bar', color='skyblue')
plt.title('Average Premium by State')
plt.ylabel('Average Premium ($)')
plt.xlabel('State')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
auto_make_by_geo = data.groupby(['Province', 'make']).size().unstack()

auto_make_by_geo.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Comparison of Auto Make by State')
plt.ylabel('Count')
plt.xlabel('State')
plt.xticks(rotation=50)

plt.legend(title='Auto Make', loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=5)  

plt.tight_layout()
plt.show()
```
